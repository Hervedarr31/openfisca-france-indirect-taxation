# Percentage of households' energy expenditures in total income

# Import general modules

```python
from __future__ import division

import pandas
import seaborn
```

# Import modules specific to OpenFisca

```python
from openfisca_france_indirect_taxation.examples.utils_example import graph_builder_line_percent
from openfisca_france_indirect_taxation.surveys import SurveyScenario
from openfisca_france_indirect_taxation.almost_ideal_demand_system.aids_estimation_from_stata import get_elasticities
from openfisca_france_indirect_taxation.examples.calage_bdf_cn_energy import get_inflators_by_year_energy
```

# Import new colors

```python
seaborn.set_palette(seaborn.color_palette("Set2", 12))
%matplotlib inline
```

# Simulate energy's expenditure share by age

```python
# Sélection des variables utilisées pour la simulation

inflators_by_year = get_inflators_by_year_energy(rebuild = False)
year = 2012
data_year = 2011
elasticities = get_elasticities(data_year)
inflation_kwargs = dict(inflator_by_variable = inflators_by_year[year])
del inflation_kwargs['inflator_by_variable']['somme_coicop12']

simulated_variables = ['depenses_energies', 'depenses_energies_logement', 'depenses_essence', 'depenses_diesel', 'poste_coicop_722',
                       'poste_coicop_451', 'poste_coicop_452', 'poste_coicop_453', 'poste_coicop_454',
                       'rev_disp_loyerimput', 'somme_coicop12']

survey_scenario = SurveyScenario.create(
    elasticities = elasticities,
    inflation_kwargs = inflation_kwargs,
    reform_key = 'taxe_carbone',
    year = year,
    data_year = data_year
    )
pivot_table = pandas.DataFrame()
for values in simulated_variables:
    pivot_table = pandas.concat([
        pivot_table,
        survey_scenario.compute_pivot_table(values = [values], columns = ['age_group_pr'])
        ])
df = pivot_table.T

df.rename(columns = {'somme_coicop12': 'total expenditures'}, inplace = True)

df['Energy share in total expenditures'] = df['depenses_energies'] / df['total expenditures']
df['Housing energy share in total expenditures'] = df['depenses_energies_logement'] / df['total expenditures']
df['Transport energy share in total expenditures'] = df['poste_coicop_722'] / df['total expenditures']
df['Diesel share in total expenditures'] = df['depenses_diesel'] / df['total expenditures']
df['Gasoline share in total expenditures'] = df['depenses_essence'] / df['total expenditures']
df['Electricity share in total expenditures'] = df['poste_coicop_451'] / df['total expenditures']
df['Natural gas share in total expenditures'] = df['poste_coicop_452'] / df['total expenditures']
df['Domestic fuel share in total expenditures'] = df['poste_coicop_453'] / df['total expenditures']
df['Solid fuels share in total expenditures'] = df['poste_coicop_454'] / df['total expenditures']

# Réalisation de graphiques
print('Percentage of energy expenditure in total expenditures')
graph_builder_line_percent(df[['Energy share in total expenditures', 'Housing energy share in total expenditures',
                               'Transport energy share in total expenditures']], 1, 1)
graph_builder_line_percent(df[['Diesel share in total expenditures', 'Gasoline share in total expenditures']], 1, 1)
graph_builder_line_percent(df[['Electricity share in total expenditures', 'Natural gas share in total expenditures',
                              'Domestic fuel share in total expenditures', 'Solid fuels share in total expenditures']], 1, 1)
```
